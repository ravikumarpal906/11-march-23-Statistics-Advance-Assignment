**Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.**

**Q1: Difference between t-test and z-test:**

- **t-test:**
  - Used when the population standard deviation is unknown.
  - Compares the means of two independent groups.
  - Example: Comparing the mean heights of two groups of students.

- **z-test:**
  - Used when the population standard deviation is known.
  - Compares the mean of a single group to a known population mean.
  - Example: Testing whether the average score on a standardized test is significantly different from the national average.

**Example scenario:**

- You are a researcher interested in comparing the average reading comprehension scores of two groups of students: those who received a new reading intervention program and those who did not. You would use a t-test because you do not know the population standard deviation for reading comprehension scores.

- You are a quality control inspector at a factory that produces widgets. You know that the population standard deviation for the weight of widgets is 0.5 ounces. You take a sample of 20 widgets and find that the average weight is 10.2 ounces. You would use a z-test to determine whether the average weight of widgets is significantly different from the known population mean of 10 ounces.


**Q2: Differentiate between one-tailed and two-tailed tests.**


**One-tailed test:**

- Used when you have a specific hypothesis about the direction of the difference between two groups or between a group and a known population mean.
- Example: You hypothesize that the new reading intervention program will lead to higher reading comprehension scores than the traditional reading program.

**Two-tailed test:**

- Used when you do not have a specific hypothesis about the direction of the difference between two groups or between a group and a known population mean.
- Example: You want to test whether the average weight of widgets is different from the known population mean of 10 ounces, but you do not know whether the average weight is higher or lower than 10 ounces.

**Example scenario:**

- You are a researcher interested in comparing the average reading comprehension scores of two groups of students: those who received a new reading intervention program and those who did not. You hypothesize that the new reading intervention program will lead to higher reading comprehension scores than the traditional reading program. You would use a one-tailed t-test because you have a specific hypothesis about the direction of the difference between the two groups.

- You are a quality control inspector at a factory that produces widgets. You know that the population standard deviation for the weight of widgets is 0.5 ounces. You take a sample of 20 widgets and find that the average weight is 10.2 ounces. You want to test whether the average weight of widgets is different from the known population mean of 10 ounces, but you do not know whether the average weight is higher or lower than 10 ounces. You would use a two-tailed z-test because you do not have a specific hypothesis about the direction of the difference between the sample mean and the population mean.

**Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.**

**Type 1 error:**

- Rejecting the null hypothesis when it is true.
- Example: Concluding that the new reading intervention program leads to higher reading comprehension scores when it does not.

**Type 2 error:**

- Failing to reject the null hypothesis when it is false.
- Example: Concluding that the average weight of widgets is not different from the known population mean when it is.

**Example scenario for Type 1 error:**

- You are a researcher interested in comparing the average reading comprehension scores of two groups of students: those who received a new reading intervention program and those who did not. You hypothesize that the new reading intervention program will lead to higher reading comprehension scores than the traditional reading program. You conduct a t-test and find that the difference between the two groups is statistically significant. However, in reality, the new reading intervention program does not lead to higher reading comprehension scores. You have made a Type 1 error.

**Example scenario for Type 2 error:**

- You are a quality control inspector at a factory that produces widgets. You know that the population standard deviation for the weight of widgets is 0.5 ounces. You take a sample of 20 widgets and find that the average weight is 10.2 ounces. You conduct a z-test and find that the difference between the sample mean and the population mean is not statistically significant. However, in reality, the average weight of widgets is different from the known population mean. You have made a Type 2 error.


**Q4: Explain Bayes's theorem with an example.**

Bayes' theorem is a fundamental concept in probability theory that describes the relationship between conditional probabilities. It is used to calculate the probability of an event occurring given prior knowledge of other related events.

The formula for Bayes' theorem is:

P(A | B) = (P(B | A) * P(A)) / P(B)

where:

- P(A | B) is the probability of event A occurring given that event B has occurred.
- P(B | A) is the probability of event B occurring given that event A has occurred.
- P(A) is the probability of event A occurring.
- P(B) is the probability of event B occurring.

**Example:**

Let's say you have a bag of 100 marbles, 50 of which are red and 50 of which are blue. You randomly select a marble from the bag without looking.

- P(red) = 0.5 (the probability of selecting a red marble)
- P(blue) = 0.5 (the probability of selecting a blue marble)

Now, let's say you reach into the bag and feel a marble. It feels slightly larger than the other marbles. You know that there is a 10% chance that a marble is slightly larger than the others.

- P(large | red) = 0.1 (the probability of selecting a marble that is slightly larger than the others given that it is red)
- P(large | blue) = 0.05 (the probability of selecting a marble that is slightly larger than the others given that it is blue)

Using Bayes' theorem, you can calculate the probability that the marble you selected is red given that it is slightly larger than the others:

P(red | large) = (P(large | red) * P(red)) / P(large)

P(red | large) = (0.1 * 0.5) / (0.1 * 0.5 + 0.05 * 0.5)

P(red | large) = 0.6667

Therefore, there is a 66.67% chance that the marble you selected is red given that it is slightly larger than the others.


**Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.**

A confidence interval is a range of values that is likely to contain the true population parameter with a certain level of confidence. It is calculated by taking the sample mean plus or minus the margin of error.

The margin of error is calculated by multiplying the critical value by the standard error of the mean. The critical value is based on the desired level of confidence and the degrees of freedom. The standard error of the mean is calculated by dividing the standard deviation by the square root of the sample size.

For example, let's say you have a sample of 100 people and you want to estimate the mean height of the population with a 95% confidence level. The sample mean is 68 inches and the sample standard deviation is 2 inches.

The critical value for a 95% confidence level with 99 degrees of freedom (100 - 1) is 1.984.

The standard error of the mean is 2 / sqrt(100) = 0.2 inches.

The margin of error is 1.984 * 0.2 = 0.3968 inches.

The 95% confidence interval is 68 - 0.3968 to 68 + 0.3968, or (67.6032, 68.3968).

This means that we are 95% confident that the true population mean height is between 67.6032 inches and 68.3968 inches.

**Q7. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5. Interpret the results.**

```python
import numpy as np
from scipy import stats
data = [51, 52, 53, 54, 55, 45,46,47,48,49]
# Define the sample mean and standard deviation
sample_mean = 50
sample_std = 5

# Calculate the margin of error
margin_error = stats.t.ppf(0.975, len(data) - 1) * sample_std / np.sqrt(len(data))

# Calculate the confidence interval
confidence_interval = (sample_mean - margin_error, sample_mean + margin_error)

# Print the confidence interval
print(confidence_interval)

# Interpret the results
print("We are 95% confident that the true population mean is between", confidence_interval[0], "and", confidence_interval[1])
```

**Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error? Provide an example of a scenario where a larger sample size would result in a smaller margin of error.**

**Margin of Error:**

The margin of error is the amount of error that is allowed in a confidence interval. It is calculated by multiplying the critical value by the standard error of the mean. The critical value is based on the desired level of confidence and the degrees of freedom. The standard error of the mean is calculated by dividing the standard deviation by the square root of the sample size.

**Effect of Sample Size on Margin of Error:**

The margin of error decreases as the sample size increases. This is because a larger sample size provides more information about the population, which makes the estimate of the population mean more precise.

**Example:**

Suppose you want to estimate the mean height of the population of adult males in the United States. You take a sample of 100 men and find that the sample mean height is 68 inches. You want to construct a 95% confidence interval for the population mean height.

The critical value for a 95% confidence level with 99 degrees of freedom (100 - 1) is 1.984. The standard error of the mean is 2 inches / sqrt(100) = 0.2 inches.

The margin of error is 1.984 * 0.2 = 0.3968 inches.

The 95% confidence interval is 68 - 0.3968 to 68 + 0.3968, or (67.6032, 68.3968).

Now, suppose you take a larger sample of 400 men and find that the sample mean height is still 68 inches. The standard error of the mean is now 2 inches / sqrt(400) = 0.1 inches.

The margin of error is 1.984 * 0.1 = 0.1984 inches.

The 95% confidence interval is 68 - 0.1984 to 68 + 0.1984, or (67.8016, 68.1984).

As you can see, the margin of error is smaller when the sample size is larger. This is because the larger sample size provides more information about the population, which makes the estimate of the population mean more precise.

**Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is significantly effective at a 95% confidence level using a t-test.**

```python
import scipy.stats as stats

# Set the significance level
alpha = 0.05

# Set the null hypothesis
null_hypothesis = 0

# Set the alternative hypothesis
alternative_hypothesis = 6

# Calculate the t-statistic
t_statistic = (6 - null_hypothesis) / (2.5 / np.sqrt(50))

# Calculate the p-value
p_value = stats.t.cdf(t_statistic, df=49)

# Make a decision
if p_value < alpha:
  print("Reject the null hypothesis. The drug is significantly effective at a 95% confidence level.")
else:
  print("Fail to reject the null hypothesis. The drug is not significantly effective at a 95% confidence level.")
```

**Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95% confidence interval for the true proportion of people who are satisfied with their job.**

```python
import numpy as np
from scipy import stats

# Calculate the sample proportion
sample_proportion = 65 / 100

# Calculate the margin of error
margin_error = 1.96 * np.sqrt((sample_proportion * (1 - sample_proportion)) / 500)

# Calculate the confidence interval
confidence_interval = (sample_proportion - margin_error, sample_proportion + margin_error)

# Print the confidence interval
print(confidence_interval)
```

**Q12. A researcher is testing the effectiveness of two different teaching methods on student performance. Sample A has a mean score of 85 with a standard deviation of 6, while sample B has a mean score of 82 with a standard deviation of 5. Conduct a hypothesis test to determine if the two teaching methods have a significant difference in student performance using a t-test with a significance level of 0.01.**

```python
# Set the significance level
alpha = 0.01

# Set the sample data
sample_a = [85] * 30
sample_b = [82] * 40

# Set the null hypothesis
null_hypothesis = 0

# Set the alternative hypothesis
alternative_hypothesis = 1

# Calculate the pooled standard deviation
pooled_std = np.sqrt(((len(sample_a) - 1) * np.var(sample_a) + (len(sample_b) - 1) * np.var(sample_b)) / (len(sample_a) + len(sample_b) - 2))

# Calculate the t-statistic
t_statistic = (np.mean(sample_a) - np.mean(sample_b)) / (pooled_std * np.sqrt(1 / len(sample_a) + 1 / len(sample_b)))

# Calculate the p-value
p_value = stats.t.cdf(t_statistic, df=len(sample_a) + len(sample_b) - 2)

# Make a decision
if p_value < alpha:
  print("Reject the null hypothesis. The two teaching methods have a significant difference in student performance.")
else:
  print("Fail to reject the null hypothesis. The two teaching methods do not have a significant difference in student performance.")
```

**Q13. A population has a mean of 60 and a standard deviation of 8. A sample of 50 observations has a mean of 65. Calculate the 90% confidence interval for the true population mean.**

```python
import numpy as np
from scipy.stats import norm

# Define the population parameters
population_mean = 60
population_std = 8

# Define the sample data
sample_mean = 65
sample_size = 50

# Calculate the standard error of the mean
standard_error = population_std / np.sqrt(sample_size)

# Calculate the critical value for a 90% confidence level
critical_value = norm.ppf(0.95)

# Calculate the confidence interval
confidence_interval = (sample_mean - critical_value * standard_error, sample_mean + critical_value * standard_error)

# Print the confidence interval
print(confidence_interval)
```

**Q14. In a study of the effects of caffeine on reaction time, a sample of 30 participants had an average reaction time of 0.25 seconds with a standard deviation of 0.05 seconds. Conduct a hypothesis test to determine if the caffeine has a significant effect on reaction time at a 90% confidence level using a t-test.**

```python
# Set the significance level
alpha = 0.1

# Set the sample data
sample_mean = 0.25
sample_std = 0.05
sample_size = 30

# Set the null hypothesis
null_hypothesis = 0

# Set the alternative hypothesis
alternative_hypothesis = 1

# Calculate the t-statistic
t_statistic = (sample_mean - null_hypothesis) / (sample_std / np.sqrt(sample_size))

# Calculate the p-value
p_value = stats.t.cdf(t_statistic, df=sample_size - 1)

# Make a decision
if p_value < alpha:
  print("Reject the null hypothesis. Caffeine has a significant effect on reaction time.")
else:
  print("Fail to reject the null hypothesis. Caffeine does not have a significant effect on reaction time.")
```
